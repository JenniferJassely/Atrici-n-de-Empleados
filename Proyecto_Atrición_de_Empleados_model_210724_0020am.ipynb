### Descripción de Variables del Dataset de Atrición de Empleados

Este dataset proporciona una visión exhaustiva de los factores relacionados con la atrición de empleados en una organización. Contiene datos detallados sobre atributos personales y profesionales de los empleados, incluyendo su nivel educativo, satisfacción laboral, historial de empleo y mucho más. Está diseñado para ayudar a identificar las razones detrás de la decisión de los empleados de permanecer o dejar su lugar de trabajo, ofreciendo insights valiosos para estrategias de retención de talento. A continuación, se detallan las variables incluidas en el dataset, cada una de las cuales aporta información crucial para el análisis de atrición.

- **Edad**:
    Edad del empleado.
    Se analizará ésta variable para determinar cómo influye la edad del empleado en la permanencia en determinado trabajo, además sirve para saber si existe corelacion entre la edad y otras variables que iremos determinando conforme analizemos los datos.

- **Atricion**:
    Indica si el empleado ha dejado la empresa (Sí o No).
    Este dato es clave, ya que es el valor que nos indica si el empleado se ha dado de baja de la empresa o no

- **Viajes_Negocio**:
    Frecuencia de viajes de negocio (Ninguno, Frecuente, Raro).
    Dato que nos servirá para saber que tanto influye si el empleado viaja , o no viaja, y en caso de que si viaje, si es frecuente o poco frecuente, y saber como se relaciona con la permanencia y o la antiguedad del empleado en la empresa.

- **Tarifa_Diaria**: Costo diario asociado al empleado.
    El costo diario asociado, queremos determinar si va directamente relacionado a su sueldo diario , o es costo que incluye (sueldo, viajes, etc)

- **Departamento**: Departamento en el que trabaja el empleado.
    Igualmente , en ésta fase incial del análisis de los datos, vamos a revisar los diferentes departamentos de la empresa y si existe alguna relacion con la permanencia del empleado, por ejemplo , suponiendo que existiera un departamento de contabilidad y uno de limpieza, poder comparar en ambos departamentos , el volumen de bajas y si es posible la frecuencia

- **Distancia_Desde_Casa**: Distancia en kilómetros desde la casa del empleado hasta el trabajo.
    Otra variable que creemos que es muy importante, queremos saber si influye en el tiempo que permanece el empleado en la empresa, de igual forma evaluandolo con diferentes variables por ejemplo, trabajo remoto + sueldo
    
- **Educacion**: Nivel de educación del empleado, codificado como:
  - 1 'Por Debajo de Colegio'
  - 2 'Colegio'
  - 3 'Licenciatura'
  - 4 'Maestría'
  - 5 'Doctorado'
- **Campo_Educacion**: Área de estudio del empleado.
- **Conteo_Empleados**: Número de empleados totales (generalmente 1 por fila).
- **Numero_Empleado**: Identificador único del empleado.
- **Satisfaccion_Ambiente**: Nivel de satisfacción con el entorno laboral, codificado de 1 (Bajo) a 4 (Muy Alto).
- **Genero**: Género del empleado (Masculino, Femenino).
- **Tarifa_Horaria**: Tarifa por hora de trabajo del empleado.
- **Involucramiento_Trabajo**: Nivel de involucramiento del empleado con su trabajo, codificado de 1 (Bajo) a 4 (Muy Alto).
- **Nivel_Trabajo**: Nivel de puesto del empleado dentro de la empresa.
- **Rol_Trabajo**: Posición específica del empleado dentro de la empresa.
- **Satisfaccion_Trabajo**: Nivel de satisfacción del empleado con su trabajo, codificado de 1 (Bajo) a 4 (Muy Alto).
- **Estado_Civil**: Estado civil del empleado (Soltero, Casado, Divorciado).
- **Ingreso_Mensual**: Ingreso mensual del empleado en moneda local.
- **Tarifa_Mensual**: Costo mensual asociado al empleado.
- **Num_Empresas_Trabajadas**: Cantidad de diferentes empresas en las que el empleado ha trabajado.
- **Mayor18**: Si el empleado es mayor de 18 años (Sí).
- **Horas_Extras**: Indica si el empleado ha trabajado horas extras (Sí o No).
- **Incremento_Salarial**: Porcentaje de incremento en el salario del último año.
- **Calificacion_Desempeno**: Calificación del desempeño del empleado, codificada de 1 (Bajo) a 4 (Sobresaliente).
- **Satisfaccion_Relacion**: Nivel de satisfacción del empleado con sus relaciones en el trabajo, codificado de 1 (Bajo) a 4 (Muy Alto).
- **Horas_Estandar**: Número de horas de trabajo estándar.
- **Nivel_Opciones_Acciones**: Nivel de opciones sobre acciones del empleado.
- **Total_Anios_Trabajo**: Total de años de experiencia laboral del empleado.
- **Capacitaciones_Ultimo_Anio**: Número de veces que el empleado recibió capacitación en el último año.
- **Equilibrio_Vida_Laboral**: Calidad del equilibrio entre la vida laboral y personal del empleado, codificado de 1 (Malo) a 4 (Excelente).
- **Anios_Compania**: Total de años que el empleado ha trabajado en la compañía.
- **Anios_Rol_Actual**: Años que el empleado ha pasado en su posición actual.
- **Anios_Ultima_Promocion**: Años transcurridos desde la última promoción del empleado.
- **Anios_Con_Gerente_Actual**: Años que el empleado ha trabajado bajo el mismo gerente.

Cargamos el dataset desde un archivo CSV y visualizamos las primeras cinco filas del dataframe

```python
def validarCSV(rutaArchivo):
    try:
        datos = pd.read_csv(rutaArchivo )
        return "Correcto"
    except FileNotFoundError:
        return "Error: El archivo no se encontró."
    except pd.errors.EmptyDataError:
        return "Error: El archivo está vacío."
    except Exception as e:
        return "Error: {e}"



def validar_columnas_csv(rutaCSV, diccionario_variables):
    datos = pd.read_csv(rutaCSV)
    columnas_faltantes = [col for col in diccionario_variables.keys() if col not in datos.columns]

    if columnas_faltantes:
        error = (f"El archivo CSV no contiene las siguientes columnas necesarias: {', '.join(columnas_faltantes)}")
        return error
    else:
        return "Correcto"


def cargaDatos(rutaCSV):
    # Cargar los datos
    datos = pd.read_csv(rutaCSV)   
    return datos

```

Usaremos el método del rango intercuartílico (IQR) para identificar valores atípicos y tratar outliers 

```python
import pandas as pd
import numpy as np

# Datos sin limpiar, sin categorizar y sin transformar valores
datos.to_csv('archivo01_no_optimizado.csv', index=False)

# Sabemos que no traemos valores nulos, sin embargo, se agrega por si más adelante nos llega otro dataset
# Identificación y tratamiento de valores faltantes
#print("Valores faltantes por columna:")
#print(datos.isnull().sum())

# Eliminar todas las filas con algún valor faltante
datos = datos.dropna()
#print("\nDespués de eliminar filas con valores faltantes:")
#print(datos.isnull().sum())

# Función para identificar outliers
def identificar_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))
    return outliers

# Función para reemplazar outliers con la mediana
def reemplazar_outliers_con_mediana(df, exclude_columns=[]):
    outliers = identificar_outliers(df)
    for column in df.columns:
        if column not in exclude_columns:
            if outliers[column].any():
                median = df[column].median()
                df[column] = np.where(outliers[column], median, df[column])
    return df

# Especificar las columnas a excluir del tratamiento de outliers
exclude_columns = ['Atricion']

# Identificar y reemplazar outliers en todas las columnas numéricas excepto las excluidas
datos_num = datos.select_dtypes(include=['number'])
datos_num = reemplazar_outliers_con_mediana(datos_num, exclude_columns)

# Reincorporar las columnas no numéricas al dataset
datos.update(datos_num)

# Verificar que no hay más outliers
outliers_despues = identificar_outliers(datos_num)
print("Número de outliers después de reemplazar con la mediana por columna:")
print(outliers_despues.sum())
```
